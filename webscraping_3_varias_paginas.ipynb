```python
# Importação das Bibliotecas
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
import requests
from requests_html import HTMLSession
import time

# Configurações de visualização do dataset
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 90)
```

```python
# Solicitação dos conteúdos HTML da página
session = HTMLSession()
response = session.get('https://www.scrapethissite.com/pages/forms/')
```

```python
# Status da solicitação
response.status_code
```

```python
# Analisar o documento HTML da página
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Visualização da página em formato HTML
print(soup.prettify())
```

```python
# URL de todas as páginas
page = 1
while page != 25:
    url = f"https://www.scrapethissite.com/pages/forms/?page_num={page}"
    print(url)
    page = page + 1
```

```python
%%time

# Web Scraping

nomes = []
anos = []
vitorias = []
derrotas = []
taxas = []
gols_pro = []
gols_contra = []
saldos_gols = []

page = 1

while page != 25:
    url = f"https://www.scrapethissite.com/pages/forms/?page_num={page}"
    time.sleep(2)
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html, "lxml")

    elementos = soup.findAll('tr', class_='team')
            
    for itens in range(len(elementos)):
        

        nome = elementos[itens].find("td", class_="name").text.replace('\n', '').strip()
        ano = elementos[itens].find('td', class_='year').text.replace('\n', '').strip()
        vitoria = elementos[itens].find('td', class_='wins').text.replace('\n', '').strip()
        derrota = elementos[itens].find('td', class_='losses').text.replace('\n', '').strip()
        taxa = elementos[itens].find('td', class_='pct text-danger').text.strip() if elementos[itens].find('td', class_='pct text-danger') else "None"
        gol_pro = elementos[itens].find('td', class_='gf').text.replace('\n', '').strip()
        gol_contra = elementos[itens].find('td', class_='ga').text.replace('\n', '').strip()
        saldo_gols = elementos[itens].find('td', class_='diff text-success').text.strip() if elementos[itens].find('td', class_='diff text-success') else "None"

        nomes.append(nome)
        anos.append(ano)
        vitorias.append(vitoria)
        derrotas.append(derrota)
        taxas.append(taxa)
        gols_pro.append(gol_pro)
        gols_contra.append(gol_contra)
        saldos_gols.append(saldo_gols)
        
    page = page + 1
```

```python
# DataFrame
df = pd.DataFrame({'Time': nomes, 'Ano': anos, 'Vitórias': vitorias, 'Derrotas': derrotas, 'Taxa de Vitórias %': taxas,
                  'Gols Pro': gols_pro, 'Gols Contra': gols_contra, 'Saldo de Gols': saldos_gols})
```

```python
df
```

```python
# Salvar o DataFrame em formato CSV
df.to_csv('webscraping_3_ex1')
```

```python
# Solicitação dos conteúdos HTML da página
session = HTMLSession()
response = session.get('http://books.toscrape.com/index.html')
```

```python
# Status da solicitação
response.status_code
```

```python
# Analisar o documento HTML da página
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Visualização da página em formato HTML
print(soup.prettify())
```

```python
# URL de todas as páginas
page = 1

while page != 51:
    
    url = f"http://books.toscrape.com/catalogue/page-{page}" + ".html"
    print(url)
    page = page + 1
```

```python
%%time

titulos = []
precos = []
estoques = []
avaliacoes = []

page = 1

while page != 51:
    
    url = f"http://books.toscrape.com/catalogue/page-{page}" + ".html"
    time.sleep(2)
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html, "lxml")

    elementos = soup.find('ol', {'class':'row'}).find_all('li')

    for itens in range(len(elementos)):

        titulo = elementos[itens].find('h3').text
        preco = elementos[itens].find('p', class_='price_color').text.replace('£', '').strip()
        estoque = elementos[itens].find('p', class_='instock availability').text.strip()
        avaliacao = elementos[itens].find('p').get('class')[1]
                                        
        titulos.append(titulo)
        precos.append(preco)
        estoques.append(estoque)
        avaliacoes.append(avaliacao)
               
    page = page + 1
```

```python
# Dataframe
df = pd.DataFrame({'Titulos': titulos, 'Preço': precos, 'Estoque': estoques, 'Avaliação': avaliacoes})
```

```python
df
```

```python
# Salvar o DataFrame em formato CSV
df.to_csv('webscraping_3_ex2')
```
